## cleaning of Super Built up area, Built Up area, Carpet area

```python
import re
df['Super Built up area']=df['Super Built up area'].str.extract(r'^(\d+)')

df['Super Built up area'] = df['Super Built up area'].astype(float)
```

```python
def clean_built_up_area(x):
    if pd.isnull(x):
        return np.nan
    try:
        return round(float(str(x).split()[0]), 2)
    except:
        return None

df['Built Up area'] = df['Built Up area'].apply(clean_built_up_area)
```

```python
def clean_built_up_area(x):
    if pd.isnull(x):
        return np.nan
    try:
        return round(float(str(x).split()[0]), 2)
    except:
        return None

df['Carpet area'] = df['Carpet area'].apply(clean_built_up_area)
```
