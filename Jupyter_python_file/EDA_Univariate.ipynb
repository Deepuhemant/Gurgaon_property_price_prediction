```python
# skewness and kurtosis
skewness = df['price_in_crores'].skew()
kurtosis = df['price_in_crores'].kurt()
print(skewness, kurtosis)
```

**skewness** : The price distribution has a skewness of approximately 3.279 indicating a positive skew. This means that the distribution tail is skewed to right, which aligns with our observation from the histogram where most properties have prices on the lower end with a few high-priced properties.

**kurtosis** : The kurtosis value is approximately 14.93. A kurtosis value greater than 3 indicates a distribution with heavier tails and more outliers compared to normal distribution.

```python
# Quantile Analysis
quantiles = df['price_in_crores'].quantile([0.01,0.05,0.95,0.99])
```

```python
quantiles
```

**Quantile Analysis :**
  + 1% Quantile : Only 1% of properties are priced below 0.25 crores.
  + 5% Quantile : 5% of properties are priced below 0.37 crores.
  + 95% Quantile : 95% of properties are priced below 8.5 crores.
  + 99% Quantile : 99% of properties are priced below 15.26 crores, indicating that very few properties are priced above this value.
  

```python
# Identify potential outliers using IQR method
Q1 = df['price_in_crores'].describe()['25%']
Q3 = df['price_in_crores'].describe()['75%']
IQR = Q3-Q1
print(IQR)
```

```python
lower_fence = Q1 - IQR*1.5
higher_fence = Q3 + IQR*1.5

print(lower_fence, higher_fence)
```

```python
outliers = df[(df['price_in_crores']>higher_fence) | (df['price_in_crores']<lower_fence)]
outliers.head()
```

```python
outliers.shape
```

```python
outliers['price_in_crores'].describe()
```

**Outliers Analysis (using IQR method)**
  + Based on the IQR method, there are 425 properties considered as outliers.
  + These outliers have an average price of approximately 9.24 crores
  + The range for these outliers is from 5.46 crores to 31.50 crores

```python
# price binning 
bins = [0,1,2,3,5,10,20,50]
bin_labels = ['0-1','1-2','2-3','3-5','5-10','10-20','20-50']
pd.cut(df['price_in_crores'], bins=bins, labels=bin_labels, right=False).value_counts().sort_index().plot(kind='bar')
```

```python
# ecdf plot
ecdf = df['price_in_crores'].value_counts().sort_index().cumsum()/len(df['price_in_crores'])
plt.plot(ecdf.index, ecdf, marker='.', linestyle='none')
plt.grid()
```

```python
plt.figure(figsize=(15,6))

#Distribution plot without log transformation
plt.subplot(1,2,1)
sns.histplot(df['price_in_crores'], kde=True, bins=50, color='skyblue')
plt.title('Distribution of Prices (Original)')
plt.xlabel('Price(in_crores)')
plt.ylabel('Frequency')

#Distribution plot using log transformation
plt.subplot(1,2,2)
sns.histplot(np.log1p(df['price_in_crores']), kde=True, bins=50, color='red')
plt.title('Distribution of Prices (Log transformed)')
plt.xlabel('Log price(in_crores)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

np.log1p(x): This function computes the natural logarithm of 1+x. it is designed to provide more accurate results for values of x that are very close to zero.
Using np.log1p helps in transforming the price column while ensuring that any value (including zero, if present) is handled appropriately. when we need to reverse the tranformation, we can use np.expm1 which computes e*pow(x)-1

```python
# skewness and kurtosis
skewness = np.log1p(df['price_in_crores']).skew()
kurtosis = np.log1p(df['price_in_crores']).kurt()
print(skewness, kurtosis)
```

```python
plt.figure(figsize=(15,6))

#Distribution plot without log transformation
plt.subplot(1,2,1)
sns.boxplot(df['price_in_crores'],  color='skyblue',orient='h')
plt.title('Distribution of Prices (Original)')
plt.xlabel('Price(in_crores)')
plt.ylabel('Frequency')

#Distribution plot using log transformation
plt.subplot(1,2,2)
sns.boxplot(np.log1p(df['price_in_crores']),  color='red',orient='h')
plt.title('Distribution of Prices (Log transformed)')
plt.xlabel('Log price(in_crores)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Built up area

```python
print(df['Built Up area'].isnull().sum())
```

```python
df['Built Up area']
```

```python
def clean_built_up_area(x):
    if pd.isnull(x):
        return np.nan
    try:
        return round(float(str(x).split()[0]), 2)
    except:
        return None

df['Built Up area'] = df['Built Up area'].apply(clean_built_up_area)
```
