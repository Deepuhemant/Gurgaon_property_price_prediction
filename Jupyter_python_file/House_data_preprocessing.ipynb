```python
# Function to extract area value based on type
def extract_area(area_type, text):
    match = re.search(r'\(([\d,]+)\s*sq\.m\.\)\s*' + re.escape(area_type), text)
    if match:
        return int(match.group(1).replace(',', ''))
    return None

# Apply the function to create new columns
df['Plot Area'] = df['area'].apply(lambda x: extract_area('Plot Area', x))
df['Built-up Area'] = df['area'].apply(lambda x: extract_area('Built-up Area', x))
df['Carpet Area'] = df['area'].apply(lambda x: extract_area('Carpet Area', x))


       
    
```
