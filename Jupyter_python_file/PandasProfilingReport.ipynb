```python
import pandas as pd
from pandas_profiling import ProfileReport

# Load your Dataset
df = pd.read_csv('gurgaon_property_aftercleaning.csv')
df= df.drop_duplicates()

# Create the profile Report object
profile_report = ProfileReport(df, title='Pandas Profiling Report', explorative=True)

# Generate the report
profile_report.to_file('output_report.html')
```
