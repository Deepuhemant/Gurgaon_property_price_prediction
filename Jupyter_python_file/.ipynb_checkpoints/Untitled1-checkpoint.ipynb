```python
import re, ast, pandas as pd

# assume df already exists and has column 'Locationadvantages'

def convert_to_km(distance_str):
    if distance_str is None:
        return None
    distance_str = str(distance_str).strip()
    match = re.match(r'([\d.]+)\s*(Meter|KM|Km|km|m|M)', distance_str, re.IGNORECASE)
    if match:
        value = float(match.group(1))
        unit = match.group(2).upper()
        if unit in ['METER', 'M']:
            return round(value / 1000, 2)
        else:
            return round(value, 2)
    return None

def convert_dict_string(cell_value, return_as_string=True):
    if pd.isna(cell_value) or str(cell_value).strip() == "":
        return cell_value
    try:
        data_dict = ast.literal_eval(cell_value)
        if not isinstance(data_dict, dict):
            return cell_value
    except Exception:
        return cell_value
    converted = {}
    for place, dist in data_dict.items():
        km_val = convert_to_km(dist)
        if km_val is None:
            converted[place] = dist
        else:
            converted[place] = f"{km_val} KM"
    return str(converted) if return_as_string else converted

# Apply to entire column (example: create a new column)
df['LocationAdvantages'] = df['LocationAdvantages'].apply(lambda x: convert_dict_string(x, return_as_string=True))
```

```python
import pandas as pd
import ast
import re

def normalize_location(name):
    """
    Normalize location name:
    - lowercase
    - remove spaces
    - remove underscores
    - remove hyphens
    - remove special chars
    """
    name = str(name).lower().strip()
    name = re.sub(r'[\s_-]+', '', name)     # remove spaces, underscores, hyphens
    name = re.sub(r'[^a-z0-9]', '', name)   # remove everything except letters & numbers
    return name

def clean_distance(value):
    """
    Remove 'km' or 'KM' or extra spaces and convert to float.
    """
    if value is None or pd.isna(value):
        return None
    
    value = str(value).lower().replace("km", "").strip()
    
    try:
        return float(value)
    except:
        return None

def parse_location_dict(cell_value):
    if pd.isna(cell_value):
        return {}

    try:
        raw_dict = ast.literal_eval(cell_value)
    except:
        return {}

    clean_dict = {}

    for k, v in raw_dict.items():
        norm_key = normalize_location(k)
        clean_value = clean_distance(v)   # convert to float
        clean_dict[norm_key] = clean_value

    return clean_dict

def process_location_advantages(df):
    # Convert each cell into dict with normalized keys + float values
    parsed = df['LocationAdvantages'].apply(parse_location_dict)

    # Expand into columns
    expanded = parsed.apply(pd.Series)

    # Merge duplicate columns created by normalization
    expanded = expanded.groupby(level=0, axis=1).first()

    # Merge with original df
    final_df = pd.concat([df, expanded], axis=1)

    return final_df
```

```python
df_processed = process_location_advantages(df)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Apply the scaler to the entire DataFrame
scaled_data3 = pd.DataFrame(scaler.fit_transform(df_processed), columns=df_processed.columns, index=df_processed.index)
```

```python
scaled_data3.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim3 = cosine_similarity(scaled_data3)
```

```python
cosine_sim3.shape
```

```python
def recommend_property_with_scores(property_name, top_n=5):

    # cosine_sim_matrix = cosine_sim1 + cosine_sim2 + 8*cosine_sim3
    cosine_sim_matrix = cosine_sim3

    #Get the similarity scores for the property using its name as the index
    sim_scores = list(enumerate(cosine_sim_matrix[scaled_data3.index.get_loc(property_name)]))

    # sort properties based on the similarity scores
    sorted_scores = sorted(sim_scores, key=lambda x : x[1], reverse=True)

    #Get the indices and scores of the top_n most similar properties
    top_indices = [i[0] for i in sorted_scores[1:top_n+1]]
    top_scores = [i[1] for i in sorted_scores[1:top_n+1]]

    #Retrieve the names of the top properties using the indices
    top_properties = scaled_data3.index[top_indices].tolist()

    # Create a DataFrame with the result
    recommendations_df = pd.DataFrame({
        'PropertyName': top_properties,
        'SimilarityScore' : top_scores
    })

    return recommendations_df

    # Test the recommender function using a property name
recommend_property_with_scores("Adani Brahma Samsara Vilasa")
```
