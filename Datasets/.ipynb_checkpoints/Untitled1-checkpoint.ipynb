```python
def recommend_properties(property_name, cosine_sim=cosine_sim1):
    
    # Get index of the property that matches the name 
    idx = df.index[df['PropertyName'] == property_name].tolist()[0]

    # Get the pairwise similarity score with that property
    sim_scores = list(enumerate(cosine_sim1[idx]))

    #Sort the property based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar properties
    sim_scores = sim_scores[1:6]

    # Get the property indices
    property_indices = [i[0] for i in sim_scores]

    recommendations_df = pd.DataFrame({
        'PropertyName' : df['PropertyName'].iloc[property_indices],
        'SimilarityScore' : sim_scores
    })

    # Return the top 10 most similar properties
    return recommendations_df
```

```python
import pandas as pd
import ast
import re


def convert_lakhs_to_crores_in_string(price_str):
    """
    Convert Lakhs (L/Lac/Lacs/Lakhs) to Crores (Cr) in a price string
    
    Examples:
    "₹ 88.45 L" -> "₹ 0.8845 Cr"
    "₹ 88.45 - 94.8 L" -> "₹ 0.8845 - 0.948 Cr"
    "₹ 88.45 Lac" -> "₹ 0.8845 Cr"
    "₹ 1.26 Cr" -> "₹ 1.26 Cr" (unchanged)
    """
    if not price_str or pd.isna(price_str):
        return price_str
    
    # Check if already in Crores
    if 'Cr' in price_str:
        return price_str
    
    # Check if it's in Lakhs (L, Lac, Lacs, Lakhs)
    lakhs_pattern = r'(L|Lac|Lacs|Lakhs)\b'
    if not re.search(lakhs_pattern, price_str, re.IGNORECASE):
        return price_str
    
    # Extract all numbers from the string
    numbers = re.findall(r'(\d+(?:\.\d+)?)', price_str)
    
    if not numbers:
        return price_str
    
    # Convert each number from Lakhs to Crores (divide by 100)
    converted_numbers = []
    for num in numbers:
        lakhs_value = float(num)
        crores_value = lakhs_value / 100
        # Round to 4 decimal places to avoid floating point issues
        crores_value = round(crores_value, 4)
        converted_numbers.append(str(crores_value))
    
    # Build the new string
    # Handle different formats
    if len(converted_numbers) == 1:
        # Single value: "₹ 88.45 L" -> "₹ 0.8845 Cr"
        new_str = f"₹ {converted_numbers[0]} Cr"
    elif len(converted_numbers) == 2:
        # Range: "₹ 88.45 - 94.8 L" -> "₹ 0.8845 - 0.948 Cr"
        new_str = f"₹ {converted_numbers[0]} - {converted_numbers[1]} Cr"
    else:
        # Fallback: just replace L/Lac/etc with Cr
        new_str = re.sub(lakhs_pattern, 'Cr', price_str, flags=re.IGNORECASE)
    
    return new_str


def convert_price_details_lakhs_to_crores(price_details_str):
    """
    Convert all Lakhs values to Crores in a PriceDetails dictionary string
    
    Args:
        price_details_str: String representation of dictionary with price details
    
    Returns:
        Modified string with all Lakhs converted to Crores
    """
    if not price_details_str or pd.isna(price_details_str):
        return price_details_str
    
    try:
        # Handle different input types
        if isinstance(price_details_str, dict):
            # Already a dictionary
            price_dict = price_details_str.copy()
        elif isinstance(price_details_str, str):
            # Try to parse as dictionary
            # First, try ast.literal_eval
            try:
                price_dict = ast.literal_eval(price_details_str)
            except (ValueError, SyntaxError):
                # If ast.literal_eval fails, try eval with restricted namespace
                # This handles cases where the string has variable names
                try:
                    import json
                    # Try to convert to valid JSON first
                    json_str = price_details_str.replace("'", '"')
                    price_dict = json.loads(json_str)
                except:
                    # If still fails, use eval with empty namespace for safety
                    price_dict = eval(price_details_str, {"__builtins__": {}}, {})
        else:
            # Unknown type, return as is
            return price_details_str
        
        # Iterate through each property type (BHK, Land, etc.)
        for property_type, details in price_dict.items():
            if isinstance(details, dict) and 'price-range' in details:
                original_price = details['price-range']
                converted_price = convert_lakhs_to_crores_in_string(original_price)
                details['price-range'] = converted_price
        
        # Convert back to string representation
        return str(price_dict)
    
    except Exception as e:
        print(f"Error converting price details: {str(e)}")
        print(f"Problematic value type: {type(price_details_str)}")
        print(f"First 200 chars: {str(price_details_str)[:200]}")
        return price_details_str


def diagnose_price_details_issues(df, price_column='PriceDetails'):
    """
    Diagnose issues in PriceDetails column to identify problematic rows
    
    Args:
        df: DataFrame to diagnose
        price_column: Column name to check
    
    Returns:
        Dictionary with diagnostic information
    """
    print("\n" + "="*100)
    print("DIAGNOSING PRICE DETAILS COLUMN")
    print("="*100 + "\n")
    
    issues = {
        'total_rows': len(df),
        'null_values': df[price_column].isna().sum(),
        'empty_strings': (df[price_column] == '').sum(),
        'dict_type': 0,
        'string_type': 0,
        'other_type': 0,
        'parseable': 0,
        'unparseable': [],
        'unparseable_samples': []
    }
    
    for idx, value in df[price_column].items():
        if pd.isna(value):
            continue
            
        # Check type
        if isinstance(value, dict):
            issues['dict_type'] += 1
            issues['parseable'] += 1
        elif isinstance(value, str):
            issues['string_type'] += 1
            # Try to parse
            try:
                ast.literal_eval(value)
                issues['parseable'] += 1
            except:
                issues['unparseable'].append(idx)
                if len(issues['unparseable_samples']) < 5:
                    issues['unparseable_samples'].append({
                        'index': idx,
                        'value': value[:200],
                        'property_name': df.loc[idx, 'PropertyName'] if 'PropertyName' in df.columns else 'N/A'
                    })
        else:
            issues['other_type'] += 1
    
    # Print summary
    print(f"Total rows: {issues['total_rows']}")
    print(f"Null values: {issues['null_values']}")
    print(f"Empty strings: {issues['empty_strings']}")
    print(f"\nData types:")
    print(f"  - Dictionary type: {issues['dict_type']}")
    print(f"  - String type: {issues['string_type']}")
    print(f"  - Other types: {issues['other_type']}")
    print(f"\nParseability:")
    print(f"  - Parseable: {issues['parseable']}")
    print(f"  - Unparseable: {len(issues['unparseable'])}")
    
    if issues['unparseable_samples']:
        print(f"\n⚠️  PROBLEMATIC ROWS (first 5):")
        print("-"*100)
        for sample in issues['unparseable_samples']:
            print(f"\nRow {sample['index']} - Property: {sample['property_name']}")
            print(f"Value (first 200 chars): {sample['value']}")
            print("-"*100)
    
    return issues


def preprocess_dataframe_prices(df, price_column='PriceDetails', diagnose=False):
    """
    Preprocess entire dataframe to convert Lakhs to Crores in PriceDetails column
    
    Args:
        df: pandas DataFrame with PriceDetails column
        price_column: name of the column containing price details (default: 'PriceDetails')
        diagnose: if True, run diagnostics before processing
    
    Returns:
        DataFrame with converted prices
    """
    # Create a copy to avoid modifying original
    df_processed = df.copy()
    
    # Run diagnostics if requested
    if diagnose:
        diagnose_price_details_issues(df_processed, price_column)
    
    # Apply conversion to each row
    print(f"\nConverting Lakhs to Crores in {len(df_processed)} rows...")
    
    df_processed[price_column] = df_processed[price_column].apply(
        convert_price_details_lakhs_to_crores
    )
    
    print("Conversion complete!")
    
    return df_processed


# ============================================================================
# USAGE EXAMPLE - HOW TO USE WITH YOUR 246 ROWS
# ============================================================================

if __name__ == "__main__":
    """
    This section only runs when you execute this file directly.
    To use with your data, import the function instead.
    """
    
    print("="*100)
    print("PRICE PREPROCESSING - USAGE INSTRUCTIONS")
    print("="*100)
    print("""
To process your 246-row dataframe, use this code in your script:

# Import the function
from price_preprocessing_corrected import preprocess_dataframe_prices

# Load your data
import pandas as pd
df = pd.read_csv('your_246_rows.csv')

# Check the shape
print(f"Input shape: {df.shape}")

# Process the data
df_processed = preprocess_dataframe_prices(df, diagnose=True)

# Check the output shape
print(f"Output shape: {df_processed.shape}")

# Save the result
df_processed.to_csv('processed_output.csv', index=False)

# The output should have 246 rows, same as input!
assert len(df_processed) == 246, "Row count mismatch!"
print("✅ All 246 rows processed successfully!")
    """)
    
```

```python
import pandas as pd
import ast
import re

class PropertyDataProcessor:
    """
    A class to process and flatten property data from PriceDetails column
    """
    
    def __init__(self, dataframe):
        """
        Initialize the processor with a dataframe
        
        Parameters:
        dataframe (pd.DataFrame): DataFrame containing PropertyName and PriceDetails columns
        """
        if dataframe is None:
            raise ValueError("Dataframe cannot be None. Please provide a valid pandas DataFrame.")
        
        if not isinstance(dataframe, pd.DataFrame):
            raise TypeError(f"Expected pandas DataFrame, got {type(dataframe)} instead.")
        
        if 'PropertyName' not in dataframe.columns or 'PriceDetails' not in dataframe.columns:
            raise ValueError("Dataframe must contain 'PropertyName' and 'PriceDetails' columns.")
        
        self.df = dataframe.copy()  # Create a copy to avoid modifying the original
        self.flattened_df = None
        
        print(f"✓ Initialized with dataframe containing {len(self.df)} rows")
        
    def extract_price_range(self, price_range_str):
        """
        Extract min and max prices from price range string as float values
        
        Parameters:
        price_range_str (str): Price range string like "₹ 2 - 2.4 Cr"
        
        Returns:
        tuple: (min_price, max_price) as floats without units
        """
        if not price_range_str or pd.isna(price_range_str):
            return None, None
        
        # Remove currency symbol and unit, extract all numbers
        cleaned = price_range_str.replace('₹', '').replace('Cr', '').replace('Lac', '').replace('L', '')
        
        # Extract all decimal numbers
        prices = re.findall(r'(\d+(?:\.\d+)?)', cleaned)
        
        if len(prices) >= 2:
            return float(prices[0]), float(prices[1])
        elif len(prices) == 1:
            return float(prices[0]), float(prices[0])
        else:
            return None, None
    
    def extract_area_value(self, area_str):
        """
        Extract a clean numeric area value (average if range, single if not)
        
        Parameters:
        area_str (str): Area string like "1,850 - 2,050 sq.ft." or "1,370 sq.ft."
        
        Returns:
        float: Single numeric area value without units
        """
        if not area_str or pd.isna(area_str):
            return None
        
        # Remove commas and extract all numbers
        area_clean = area_str.replace(',', '')
        numbers = re.findall(r'(\d+(?:\.\d+)?)', area_clean)
        
        if len(numbers) >= 2:
            # If range, return average
            return (float(numbers[0]) + float(numbers[1])) / 2
        elif len(numbers) == 1:
            return float(numbers[0])
        else:
            return None
    
    def extract_area_range(self, area_str):
        """
        Extract min and max area from area string as float values
        
        Parameters:
        area_str (str): Area string like "1,850 - 2,050 sq.ft." or "1,370 sq.ft."
        
        Returns:
        tuple: (min_area, max_area) as floats without units
        """
        if not area_str or pd.isna(area_str):
            return None, None
        
        # Remove commas and extract all numbers
        area_clean = area_str.replace(',', '')
        numbers = re.findall(r'(\d+(?:\.\d+)?)', area_clean)
        
        if len(numbers) >= 2:
            return float(numbers[0]), float(numbers[1])
        elif len(numbers) == 1:
            return float(numbers[0]), float(numbers[0])
        else:
            return None, None
    
    def flatten_price_details(self):
        """
        Flatten the PriceDetails column into separate columns for each property type
        
        Returns:
        pd.DataFrame: Flattened dataframe
        """
        if self.df is None:
            raise ValueError("Dataframe is None. Cannot process.")
        
        # Create a list to store all flattened rows
        flattened_data = []
        
        for idx, row in self.df.iterrows():
            flat_row = {'PropertyName': row['PropertyName']}
            
            # Parse the PriceDetails string to dictionary
            try:
                if pd.isna(row['PriceDetails']) or row['PriceDetails'] == '':
                    flattened_data.append(flat_row)
                    continue
                    
                # Convert string representation to actual dictionary
                price_dict = ast.literal_eval(row['PriceDetails'])
                
                # Iterate through each property type (BHK, Land, etc.)
                for property_type, details in price_dict.items():
                    # Normalize property type to lowercase and remove spaces for column naming
                    # For "2 BHK" -> "2_bhk", for "Land" -> "land"
                    prop_suffix = property_type.lower().replace(' ', '_')
                    
                    # Extract building_type
                    if 'building_type' in details:
                        building_val = details['building_type']
                        # If empty string, use None
                        flat_row[f'{prop_suffix}_building_type'] = building_val if building_val else None
                    
                    # Extract area_type
                    if 'area_type' in details:
                        flat_row[f'{prop_suffix}_area_type'] = details['area_type']
                    
                    # Extract area as clean numeric value (without units)
                    if 'area' in details:
                        area_numeric = self.extract_area_value(details['area'])
                        flat_row[f'{prop_suffix}_area'] = area_numeric
                        
                        # Extract and split area into min and max (numeric values)
                        min_area, max_area = self.extract_area_range(details['area'])
                        flat_row[f'{prop_suffix}_area_min'] = min_area
                        flat_row[f'{prop_suffix}_area_max'] = max_area
                    
                    # Extract and split price range (numeric values)
                    if 'price-range' in details:
                        min_price, max_price = self.extract_price_range(details['price-range'])
                        flat_row[f'{prop_suffix}_price_range_min'] = min_price
                        flat_row[f'{prop_suffix}_price_range_max'] = max_price
                
                flattened_data.append(flat_row)
                
            except Exception as e:
                print(f"Error processing row {idx}: {e}")
                print(f"PriceDetails content: {row['PriceDetails']}")
                flattened_data.append(flat_row)
        
        # Create new dataframe
        self.flattened_df = pd.DataFrame(flattened_data)
        
        return self.flattened_df
    
    def display_preview(self, num_rows=5):
        """
        Display a preview of the flattened dataframe
        
        Parameters:
        num_rows (int): Number of rows to display
        """
        if self.flattened_df is None:
            print("No flattened data available. Please run flatten_price_details() first.")
            return
        
        print("=" * 80)
        print("PREVIEW OF FLATTENED DATA")
        print("=" * 80)
        print(f"\nShowing first {num_rows} rows:\n")
        print(self.flattened_df.head(num_rows))
        print(f"\nTotal rows: {len(self.flattened_df)}")
        print(f"Total columns: {len(self.flattened_df.columns)}")
    
    def display_columns(self):
        """
        Display all column names in the flattened dataframe
        """
        if self.flattened_df is None:
            print("No flattened data available. Please run flatten_price_details() first.")
            return
        
        print("=" * 80)
        print("ALL COLUMNS IN FLATTENED DATA")
        print("=" * 80)
        for i, col in enumerate(self.flattened_df.columns, 1):
            print(f"{i}. {col}")
        print(f"\nTotal columns: {len(self.flattened_df.columns)}")
    
    def save_to_csv(self, filename='flattened_property_data.csv', index=False):
        """
        Save the flattened dataframe to a CSV file
        
        Parameters:
        filename (str): Name of the output CSV file
        index (bool): Whether to include index in CSV
        
        Returns:
        str: Path to saved file
        """
        if self.flattened_df is None:
            print("No flattened data available. Please run flatten_price_details() first.")
            return None
        
        try:
            self.flattened_df.to_csv(filename, index=index)
            print("=" * 80)
            print("FILE SAVED SUCCESSFULLY")
            print("=" * 80)
            print(f"File saved as: {filename}")
            print(f"Total rows saved: {len(self.flattened_df)}")
            print(f"Total columns saved: {len(self.flattened_df.columns)}")
            return filename
        except Exception as e:
            print(f"Error saving file: {e}")
            return None
    
    def get_statistics(self):
        """
        Display statistics about the flattened data
        """
        if self.flattened_df is None:
            print("No flattened data available. Please run flatten_price_details() first.")
            return
        
        print("=" * 80)
        print("DATA STATISTICS")
        print("=" * 80)
        
        # Get numeric columns
        numeric_cols = self.flattened_df.select_dtypes(include=['float64', 'int64']).columns
        
        print(f"\nNumeric columns: {len(numeric_cols)}")
        print(f"Text columns: {len(self.flattened_df.columns) - len(numeric_cols)}")
        
        print(f"\nMissing values per column:")
        missing_counts = self.flattened_df.isnull().sum()
        missing_counts = missing_counts[missing_counts > 0]
        if len(missing_counts) > 0:
            print(missing_counts)
        else:
            print("No missing values!")
        
        # Show data types
        print(f"\nData types:")
        print(self.flattened_df.dtypes.value_counts())
    
    def process_all(self, csv_filename='flattened_property_data.csv'):
        """
        Run all processing steps in sequence
        
        Parameters:
        csv_filename (str): Name of the output CSV file
        
        Returns:
        pd.DataFrame: The flattened dataframe
        """
        print("Starting data processing...\n")
        
        # Step 1: Flatten the data
        print("Step 1: Flattening price details...")
        self.flatten_price_details()
        print("✓ Flattening complete\n")
        
        # Step 2: Display preview
        print("Step 2: Displaying preview...")
        self.display_preview()
        print("\n")
        
        # Step 3: Display columns
        print("Step 3: Displaying all columns...")
        self.display_columns()
        print("\n")
        
        # Step 4: Display statistics
        print("Step 4: Displaying statistics...")
        self.get_statistics()
        print("\n")
        
        # Step 5: Save to CSV
        print("Step 5: Saving to CSV...")
        self.save_to_csv(csv_filename)
        print("\n")
        
        print("=" * 80)
        print("ALL PROCESSING COMPLETE!")
        print("=" * 80)
        
        return self.flattened_df


# ============================================================================
# USAGE EXAMPLE
# ============================================================================

if __name__ == "__main__":
    # Sample data for demonstration
    sample_data = {
        'PropertyName': [
            'Smartworld One DXP',
            'M3M Crown',
            'Adani Brahma Samsara Vilasa',
            'Sobha City',
            'Signature Global City 93',
            'Premium Land Plot'
        ],
        'PriceDetails': [
            "{'2 BHK': {'building_type': 'Apartment', 'area_type': 'Carpet Area', 'area': '1,370 sq.ft.', 'price-range': '₹ 2 - 2.4 Cr'}, '3 BHK': {'building_type': 'Apartment', 'area_type': 'Carpet Area', 'area': '1,850 - 2,050 sq.ft.', 'price-range': '₹ 2.25 - 3.59 Cr'}, '4 BHK': {'building_type': 'Apartment', 'area_type': 'Carpet Area', 'area': '2,600 sq.ft.', 'price-range': '₹ 3.24 - 4.56 Cr'}}",
            "{'3 BHK': {'building_type': 'Apartment', 'area_type': 'Super Area', 'area': '2,000 sq.ft.', 'price-range': '₹ 3 - 4 Cr'}}",
            "{'3 BHK': {'building_type': 'Independent Floor', 'area_type': 'Built-up Area', 'area': '1,800 sq.ft.', 'price-range': '₹ 2.5 - 3.2 Cr'}}",
            "{'2 BHK': {'building_type': 'Apartment', 'area_type': 'Carpet Area', 'area': '1,200 sq.ft.', 'price-range': '₹ 1.8 - 2.2 Cr'}, '3 BHK': {'building_type': 'Apartment', 'area_type': 'Carpet Area', 'area': '1,600 sq.ft.', 'price-range': '₹ 2.5 - 3 Cr'}}",
            "{'2 BHK': {'building_type': 'Independent Floor', 'area_type': 'Super Area', 'area': '1,100 sq.ft.', 'price-range': '₹ 1.5 - 1.8 Cr'}}",
            "{'Land': {'building_type': '', 'area_type': 'Plot Area', 'area': '2,160 - 3,240 sq.ft.', 'price-range': '₹ 4.2 - 6.3 Cr'}}"
        ]
    }
    
    # Create dataframe
    df = pd.DataFrame(sample_data)
    
    print("ORIGINAL DATA:")
    print(df)
    print("\n" + "=" * 80 + "\n")
    
    # Create an instance of PropertyDataProcessor
    processor = PropertyDataProcessor(df)
    
    # Process all steps at once
    result_df = processor.process_all(csv_filename='property_data_flattened.csv')
    
    # Show specific numeric columns
    print("\n" + "=" * 80)
    print("SAMPLE NUMERIC VALUES")
    print("=" * 80)
    
    numeric_cols = result_df.select_dtypes(include=['float64', 'int64']).columns
    if len(numeric_cols) > 0:
        print("\nFirst row numeric values:")
        for col in numeric_cols[:10]:  # Show first 10 numeric columns
            print(f"  {col}: {result_df[col].iloc[0]}")
```
